```python
# Imprimir números del -100 al 200
for i in range(-100, 201):
    print(i)
```

```python
# Imprimir números del -100 al 200 que son divisibles entre 6
for i in range(-100, 201):
    if i % 6 == 0:
        print(i)
```

```python
# Función para sumar dos números y mostrar un mensaje dependiendo del resultado
def sumador(num1, num2):
    resultado = num1 + num2

    print("El resultado de la suma es:", resultado)

    if resultado < 200:
        print("Menor a 200")
    elif resultado > 200 and resultado < 250:
        print("Mayor a 200")
    else:
        print("Mayor a 250")

# Solicitar al usuario que ingrese dos números
numero1 = float(input("Ingrese el primer número: "))
numero2 = float(input("Ingrese el segundo número: "))

# Llamar a la función sumador con los números ingresados por el usuario
sumador(numero1, numero2)
```

```python
# Definir una función para evaluar la edad y preferencia musical del usuario
def evaluar_usuario():
    nombre = input("¿Cuál es tu nombre? ")
    edad = int(input("¿Cuál es tu edad? "))
    gusta_musica_urbana = input("¿Te gusta la música urbana? (Sí/No) ").lower()

    if edad >= 18:
        if gusta_musica_urbana == "sí":
            print(f"¡Hola {nombre}! Eres mayor de edad y te gusta la música urbana. ¡Qué genial!")
        elif gusta_musica_urbana == "no":
            print(f"¡Hola {nombre}! Eres mayor de edad, pero no te gusta la música urbana. ¡Eso está bien, cada uno tiene sus gustos!")
        else:
            print("Respuesta inválida. Por favor, responde 'Sí' o 'No' a la pregunta sobre la música urbana.")
    else:
        print(f"Lo siento {nombre}, este programa es solo para mayores de edad.")

# Llamar a la función evaluar_usuario para comenzar la evaluación
evaluar_usuario()
```
